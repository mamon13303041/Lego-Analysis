**datasets/lego_sets.csv**

set_num: A code that is unique to each set in the dataset. This column is critical, and a missing value indicates the set is a duplicate or invalid!
set_name: A name for every set in the dataset (note that this can be the same for different sets).
year: The date the set was released.
num_parts: The number of parts contained in the set. This column is not central to our analyses, so missing values are acceptable.
theme_name: The name of the sub-theme of the set.
parent_theme: The name of the parent theme the set belongs to. Matches the `name` column of the `parent_themes` csv file.


**datasets/parent_themes.csv**

id: A code that is unique to every theme.
name: The name of the parent theme.
is_licensed: A Boolean column specifying whether the theme is a licensed theme.

```python
import pandas as pd
```

```python
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
parent_theme.to_csv("parent_them.csv")
parent_theme.head()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')

df.to_csv('logo.csv')
df.head()
```

```python
merge=df.merge(parent_theme,left_on="parent_theme", right_on="name")
merge.drop(columns='name_y', inplace=True)
merge.head(50)
```

Drop NaN value is set_name column

```python
from pandas.io.formats.style import Subset
merge[merge['set_num'].isnull()].shape
merge.dropna(subset=['set_num'])
```

 Task #1: What percentage of all licensed sets ever released were Star Wars Themed?
Percentage sould me integer

```python
licesed_df=merge[merge['is_licensed']]
licesed_df=licesed_df.dropna(subset=['set_num'])
licesed_df
```

```python
Star_Wars_Themed=licesed_df[licesed_df['theme_name']=='Star Wars']
percent_tage=int(Star_Wars_Themed.shape[0]/licesed_df.shape[0]*100)
percent_tage
```

Task #2: In which year was Star Wars not the most popular licensed theme?


```python
sort_licensed_stwar=Star_Wars_Themed.sort_values('year')
sort_licensed_stwar['Count Year']=1

sm_df=sort_licensed_stwar.groupby(['year','theme_name']).sum()
sm_df.sort_values(by='year',ascending=False).drop_duplicates()
```

How many unique sets were released each year (1955-2017)

```python
clean_set_df=merge[merge['set_num'].isnull()]
clean_set_df['count'] = 1
set_each_year=clean_set_df.groupby(['year']).sum().reset_index()[['year','count']]

for index, row in set_each_year.iterrows():
  print(row['year'],row['count'])
 
```
